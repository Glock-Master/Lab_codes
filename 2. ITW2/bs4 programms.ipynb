### Example: Extract all the quotes from a Website (Uses html)

```python
# Fetch the html content
url = 'https://www.brainyquote.com/top_100_quotes'
doc = requests.get(url).content
soup = bs.BeautifulSoup(doc)
```

```python
# Analyze where are quotes
print(soup.prettify())
```

```python
# extract quotes and author name
quotes = []
for i in soup.select("[class=block-quote]"):
    quotes.append((i.text, i.find_next(class_ = 'block-author').text))
```

```python
for i in quotes:
    print(i[0], i[1])
    print('-------------------------------------------------------------')
```

### Program 1: Extract a table (Uses html)

```python
# HTML table: 
# <table>
#     <tbody>
#         <tr>
#             <td> string </td>
#         </tr>
#     </tbody>
# </table>
```

```python
# Fetch html content
url = 'https://pythonprogramming.net/parsememcparseface/'
soup = make_soup(requests.get(url).content, 'html5lib')
table = soup.table
```

```python
# create list of data of each row
row = [[i.text for i in row_.find_all('td')] for row_ in table.find_all('tr')]
for i in row:
    print(i)
```

```python
# adding a new row
new_row = soup.new_tag('tr')
new_data = soup.new_tag('td')
new_data.string = "New data"
new_row.append(new_data)
table.tbody.append(new_row)
```

```python
# Again fetch data from each row, as visible new data has been added to output
row = [[i.text for i in row_.find_all('td')] for row_ in table.find_all('tr')]
for i in row:
    print(i)
```

```python
print(table.prettify())
```

### Program 2: Link Extraction from Sitemaps (Uses xml) and then extract information from links (Uses html)

```python
# Fetch content from washingtonpost's sitemap
url = 'https://www.washingtonpost.com/news-sitemaps/index.xml'
doc = requests.get(url)
soup = make_soup(doc.content, 'xml')
```

```python
soup
```

```python
#Extract all the links
for i in soup.find_all('loc'):
    print(*i.stripped_strings)
```

```python
# From each link, find the strings which strat with the string 'Trump'
k = soup.find_all('loc')
for i in k:
    url = i.stripped_strings
    doc = requests.get(*url).content
    soup_temp = make_soup(doc)
    news_list = [i for i in soup_temp.find_all(string = re.compile('Trump'))]
    for j in news_list:
        print(j)
```

```python
i = k[9]
url = i.stripped_strings
doc = requests.get(*url).content
soup_temp = make_soup(doc)
news_list = [i for i in soup_temp.find_all(string = re.compile('United States'))]
for j in news_list:
    print(j, end = '\n\n')
```

```python
# 1. WAP to count the number of digits in any number.
a = int(input("Enter the number: "))
count = 0
while a > 0:
    count += 1
    a = a//10
print("Count of digits: ", count)
```

```python
# 2. WAP to print sum and product the digits in any number.
a = int(input("Enter the number: "))
sum_digit = 0
product_digit = 1
while a > 0:
    temp = a%10
    sum_digit += temp
    product_digit *= temp
    a = a//10
print("Sum of digits: ", sum_digit, "  Product of digits: ", product_digit)
```

```python
# 3. WAP to read a number and print it in reverse.
a = int(input("Enter the number: "))
reversed_a = 0
while a > 0:
    reversed_a = reversed_a * 10 + a%10
    a = a//10
print("Reversed number: ", reversed_a)
```

```python
# 4. WAP to print Fibonacci series upto given number
a = int(input("Enter the number: "))
temp_1 = 0
temp_2 = 1
for i in range(a):
    print(temp_2)
    temp_3 = temp_2
    temp_2 = temp_2 + temp_1;
    temp_1 = temp_3
```

```python
# 5. WAP to print natural number in reverse starting from given number.
a = int(input("Enter the number: "))
for i in range(a, 0, -1):
    print(i)
```

```python
#6. WAP to print multiplication table of a given number.
a = int(input("Enter the number: "))
for i in range(1, 11):
    print(a, " * ", i, " = ", a * i)
```
